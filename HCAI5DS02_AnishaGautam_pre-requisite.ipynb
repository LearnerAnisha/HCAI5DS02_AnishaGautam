1. Correlation Analysis:
• Compute Pearson correlation coefficients between MedHouseVal and all other numerical features.
• Identify which features have the strongest positive and negative correlations with house value.
• Interpret these relationships.

```python
# Select numerical columns excluding MedHouseVal itself
numerical_cols = df.select_dtypes(include=['number']).columns.drop('MedHouseVal')

# Compute Pearson correlation of MedHouseVal with other numerical features
correlations = df[numerical_cols].corrwith(df['MedHouseVal'])

# Find strongest positive and negative correlations
strongest_positive = correlations.idxmax(), correlations.max()
strongest_negative = correlations.idxmin(), correlations.min()

print("Pearson correlations with MedHouseVal:")
print(correlations)

print(f"\nStrongest positive correlation: {strongest_positive[0]} = {strongest_positive[1]:.3f}")
print(f"Strongest negative correlation: {strongest_negative[0]} = {strongest_negative[1]:.3f}")
```

2. Handling Missing Data:
• Randomly set 5% of AveRooms values to NaN (simulate missingness).
• Impute missing values using median imputation.
• Visualize and compare distributions of AveRooms before and after imputation using histograms or
boxplots.
• Discuss the effect of imputation on data distribution.

```python
# Randomly set 5% of AveRooms values to NaN
np.random.seed(42)  # for reproducibility
missing_indices = df.sample(frac=0.05).index
df.loc[missing_indices, 'AveRooms'] = np.nan

# Save original distribution before imputation for comparison
original_ave_rooms = df['AveRooms'].copy()

# Impute missing values with median
median_value = df['AveRooms'].median()
df['AveRooms'].fillna(median_value, inplace=True)

# Visualize before and after imputation
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(original_ave_rooms.dropna(), kde=True, color='blue')
plt.title('AveRooms Distribution Before Imputation')

plt.subplot(1, 2, 2)
sns.histplot(df['AveRooms'], kde=True, color='green')
plt.title('AveRooms Distribution After Imputation')

plt.tight_layout()
plt.show()
```

Discussion:
Before imputation, the distribution reflects actual data with some missing points excluded.

After imputation, missing values are replaced with the median, which may cause a slight peak at the median value.

Median imputation preserves the central tendency and is robust to outliers.

However, it may reduce variability and create an artificial spike in the distribution, potentially affecting analyses sensitive to data spread.

2. Vectorization vs Loops
• Write a function to compute element-wise square of an array using:
– a for-loop
– NumPy vectorized operation
• Compare their execution time using %%timeit or time module.

```python
# Using a for-loop
def square_with_loop(arr):
    result = []
    for val in arr:
        result.append(val ** 2)
    return np.array(result)

# Using NumPy vectorization
def square_vectorized(arr):
    return arr ** 2
```

```python
import time

array = np.random.randint(0, 100, size=1000000)

# --- For loop-based ---
start_time = time.time()
loop_result = square_with_loop(array)
loop_duration = time.time() - start_time
print(f"For-loop time: {loop_duration:.6f} seconds")

# --- For vectorized ---
start_time = time.time()
vectorized_result = square_vectorized(array)
vectorized_duration = time.time() - start_time
print(f"Vectorized time: {vectorized_duration:.6f} seconds")

# ✅ Check if results are the same
print("Results are equal:", np.array_equal(loop_result, vectorized_result))
```
